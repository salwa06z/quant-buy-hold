```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
# parameter mudah diganti
tickers = ["AAPL", "SPY"]         # Apple vs S&P 500 ETF
start = "2015-01-01"              # tanggal awal
end = None                        # None = sampai data terbaru


# 1. # Ambil data saham
raw = yf.download(tickers, start=start, end=end, auto_adjust=False, progress=False)
print("Struktur kolom raw:\n", raw.columns)
raw.head()


# Ambil hanya kolom 'Adj Close'
# Jika kolomnya MultiIndex (umum saat >1 ticker)
if isinstance(raw.columns, pd.MultiIndex):
    data = raw["Adj Close"].copy()
# Jika bukan MultiIndex (jarang untuk 2 ticker, tapi kita antisipasi)
elif "Adj Close" in raw.columns:
    # kasus 1 ticker biasanya: kolom tunggal 'Adj Close' -> kita ubah namanya ke ticker pertama
    data = raw[["Adj Close"]].rename(columns={"Adj Close": tickers[0]}).copy()
else:
    raise ValueError("Kolom 'Adj Close' tidak ditemukan di data yang diunduh.")

print("Kolom setelah pilih 'Adj Close':", list(data.columns))
data.head()


# buang baris yang ada NaN di salah satu ticker
data = data.dropna(how="any")

# pastikan urut berdasarkan tanggal
data = data.sort_index()

print(data.isna().sum())  # harusnya semua 0
data.tail(3)


# 2. Normalisasi harga
# normalisasi (mulai dari 100) supaya gampang dibandingkan
normalized = data / data.iloc[0] * 100

# 3. Hitung total return
returns = (data.iloc[-1] / data.iloc[0] - 1) * 100

# 4. Plot hasil Buy & Hold
plt.figure(figsize=(10,6))
for t in data.columns:
    plt.plot(normalized.index, normalized[t], label=t)

plt.title("Buy & Hold: AAPL vs S&P 500 (index=100 di start)")
plt.xlabel("Tanggal")
plt.ylabel("Nilai (Index = 100)")
plt.legend()
plt.grid(True)
plt.show()

print("Total Return sejak", data.index[0].date(), "sampai", data.index[-1].date())
for t in data.columns:
    print(f"{t}: {returns[t]:.2f}%")

```
